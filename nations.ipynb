```python
# Respuesta pregunta 3, Â¿Cuantos paises tienen indices de CO2 mas altos que el promedio?
# Con el metodo where se filtran los paises que tienen indices de CO2 mas altos que el promedio y mas bajos tambien
df_nations['co2_record'] = np.where(df_nations['co2'] > df_nations['co2'].mean(), 1, 0)
df_nations.sample(7)
```

```python
df_nations['co2_record'].value_counts()
# Podemos observar que tenemos 130 casos donde las emiciones de co2 son menores que el promedio y 64 casos donde son mayores que el promedio
```

```python
# para saber solo los casos donde el promedio es superior a la media puedo usar slice
df_nations['co2_record'].value_counts()[1]
```

```python
# Con el signo % entre los parentesis puedo ver el porcentaje de value_counts(%)
df_nations['co2_record'].value_counts('%')
```

### Heatmap de correlaciones

```python
df_corr = df_nations
df_corr.sample(7)
```
