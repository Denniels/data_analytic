```python
# Respuesta pregunta 2, ¿Cuantos datos tenemos en cada region?
# Para saber cuantos datos tenemos en cada region, vamos a usar la funcion .value_counts()
# value_counts() ordena los datos de mayor a menor y los imprime en forma de tabla.
df_nations.value_counts('region')
```

```python
# otra forma de hacer conteo es con el metodo de agrupacion groupby
df_nations.groupby(['region'])[['country']].count()
```

```python
# Respuesta pregunta 3, ¿Cuantos paises tienen indices de CO2 mas altos que el promedio?
# Con el metodo where se filtran los paises que tienen indices de CO2 mas altos que el promedio y mas bajos tambien
df_nations['co2_record'] = np.where(df_nations['co2'] > df_nations['co2'].mean(), 1, 0)
df_nations.sample(7)
```

```python
df_nations['co2_record'].value_counts()
# Podemos observar que tenemos 130 casos donde las emiciones de co2 son menores que el promedio y 64 casos donde son mayores que el promedio
```

```python
# para saber solo los casos donde el promedio es superior a la media puedo usar slice
df_nations['co2_record'].value_counts()[1]
```

```python
# Con el signo % entre los parentesis puedo ver el porcentaje de value_counts(%)
df_nations['co2_record'].value_counts('%')
```

```python
# Respuesta pregunta 4 ¿que se puede decir del alfavetismo en africa o europa?
# Realizaremos un fuiltro para responder esta pregunta
df_nations[df_nations['region'] == 'Africa']
```

```python
df_africa = df_nations[df_nations['region'] == 'Africa']
df_africa.sample(7)
```

```python
df_europa = df_nations[df_nations['region'] == 'Europe']
df_europa.sample(7)
```

```python
df_africa['literacy'].mean()
```

```python
df_europa['literacy'].mean()
```
